```python
#Import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Setting parameters
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(10,5))
#ax.scatter(x,y)
#ax.bar(x,y)
#ax.barh(x,y)
#ax.plot(x,y)
ax.hist(y)
```

```python
#Setting fonts
font1 = {'family': 'sans-serif', 'color': 'blue', 'size':20}
font2 = {'family': 'monospace', 'color': 'green', 'size':14}

#Developing plot
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,y)
#ax.hist(y)

#Setting graphics
ax.set_title('Total growth over time', fontdict = font1, loc = 'center')
ax.set_ylabel('Total Growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
#Developing plots
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x,
           y,
           marker = 'D',
           color = 'indigo',)
```

```python
#Developing plots, including colors and markers
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y,
        marker = "*",
        color = '#7425B9',
        linestyle = '--',
        linewidth = 2,
        markersize = 12,
        markeredgecolor = '#FA9359',
        markerfacecolor = '#000000') 
ax.grid(axis = 'y', 
        color = 'red',
        linewidth = 3)
```

```python
#Import necessary packages
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Setting themes
sns.set_theme(style="darkgrid")
```

```python
#Setting tips
tips = sns.load_dataset("tips")
```

```python
#Produce graphics
sns.barplot(
    x="total_bill",
    y="day",
    data=tips,
    estimator=sum,
    ci=None,
    color='#69b3a2'
);
```
