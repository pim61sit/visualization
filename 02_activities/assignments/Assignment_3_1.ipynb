```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Industry': [
        'Agriculture, forestry, fishing and hunting [11]',
        'Mining, quarrying, and oil and gas extraction [21]',
        'Construction [23]',
        'Manufacturing [31-33]',
        'Wholesale trade [41]',
        'Retail trade [44-45]',
        'Transportation and warehousing [48-49]',
        'Information and cultural industries [51]',
        'Finance and insurance [52]',
        'Real estate and rental and leasing [53]',
        'Professional, scientific and technical services [54]',
        'Administrative and support, waste management and remediation services [56]',
        'Educational services [61]',
        'Health care and social assistance [62]',
        'Arts, entertainment and recreation [71]',
        'Accommodation and food services [72]',
        'Other services (except public administration) [81]',
        'Public administration [91]'
    ],
    'Unemployment_Rate': [5.8, 4.4, 8.9, 4.2, 3.6, 4.5, 4.5, 5.5, 6.0, 1.7, 3.6, 8.8, 2.6, 1.5, 10.5, 6.0, 7.0, 8.0],
    'Vacancy_Rate': [4.0, 2.1, 2.1, 1.7, 2.5, 1.4, 2.1, 1.7, 2.2, 2.1, 2.9, 2.9, 1.2, 4.4, 2.6, 3.3, 3.8, 2.2]
}
```

```python
df = pd.DataFrame(data)
print(df.head())
```

```python
#plotting

plt.figure(figsize=(10,7))
sns.scatterplot(
    data=df,
    x='Unemployment_Rate',
    y='Vacancy_Rate',
    hue='Industry',
    s=100,
    palette='tab20',
    legend=False
    )

#Annotate each point
for i in range(df.shape[0]):
    plt.text(
        df.Unemployment_Rate[i] +0.1,
        df.Vacancy_Rate[i],
        df.Industry[i],
        fontsize=8
    )

# Labels and titles
plt.title('Unemployment and Job Vacancy Rates by Industry (Q1 2025)', fontsize=14)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Vacancy Rate(%)')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
